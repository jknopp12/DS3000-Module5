## Question 1:

5 pts) import the data into a dataframe, and extract the following columns: Broad Phase of Flight, Total Uninjured, Total Minor Injuries, Total Serious Injuries, Total Fatal Injuries, Air Carrier, Make, Model, Aircraft Category, Aircraft Damage, Injury Severity, Country, Event Date, Investigation Type, Event Id.

Remove leading and trailing white-spaces from the selected columns, convert their data to suitable data types and perform at least one additional data preparation step.

### Prepping the Dataframe
This includes reading the csv file, stripping the columns and data of leading and trailing whitespaces, and extracting the desired columns.

```python
# read the csv file and load into dataframe
df = pd.read_csv("aviation_data.csv", delimiter='|')

# strip the leading and trailing whitespace from the column names
df.columns = [col.strip() for col in df.columns]

# create a variable for the columns to be extracted as stated in the problem
columns_to_extract = [
    "Broad Phase of Flight",
    "Total Uninjured",
    "Total Minor Injuries",
    "Total Serious Injuries",
    "Total Fatal Injuries",
    "Air Carrier",
    "Make",
    "Model",
    "Aircraft Category",
    "Aircraft Damage",
    "Injury Severity",
    "Country",
    "Event Date",
    "Investigation Type",
    "Event Id"
]

# update the dataframe to contain the extracted columns
df = df[columns_to_extract]

# get rid of leading and trailing whitespaces for the data in the dataframe
df["Broad Phase of Flight"] = df["Broad Phase of Flight"].str.strip()
df["Total Uninjured"] = df["Total Uninjured"].str.strip()
df["Total Minor Injuries"] = df["Total Minor Injuries"].str.strip()
df["Total Serious Injuries"] = df["Total Serious Injuries"].str.strip()
df["Total Fatal Injuries"] = df["Total Fatal Injuries"].str.strip()
df["Air Carrier"] = df["Air Carrier"].str.strip()
df["Make"] = df["Make"].str.strip()
df["Model"] = df["Model"].str.strip()
df["Aircraft Category"] = df["Aircraft Category"].str.strip()
df["Aircraft Damage"] = df["Aircraft Damage"].str.strip()
df["Injury Severity"] = df["Injury Severity"].str.strip()
df["Country"] = df["Country"].str.strip()
df["Event Date"] = df["Event Date"].str.strip()
df["Investigation Type"] = df["Investigation Type"].str.strip()
df["Event Id"] = df["Event Id"].str.strip()
```

### Handling Missing or Invalid Values
Here, I noticed that empty values could be replaced with unknown in the 'Broad Phase of Flight' column since there were other existing unknown values. For the rest of the dataframe I searched for whitespaces and filled them with NaN.

```python
# Make the empty values to be unknown for 'Broad Phase of Flight'
df['Broad Phase of Flight'] = df['Broad Phase of Flight'].replace('', "UNKNOWN")

# wherever there is a whitespace in the rest of the dataframe fill with NaN
columns = df.columns
# loop through all values in dataframe
for c in columns:
    # wherever there is a whitespace, replace with NaN
    df[c] = df[c].replace('', np.nan)
```

### Converting to Suitable Datatypes
In this step, the 'Event Date' column had to be converted to datatype datetime. The other modification made was converting 'Total Unijured', 'Total Minor Injuries', 'Total Serious Injuries', and 'Total Fatal Injuries' to an integer type.

```python
# change the datatype of Event Date to datetime
df["Event Date"] = pd.to_datetime(df['Event Date'], format='%m/%d/%Y')

int_cols = ['Total Uninjured', 'Total Minor Injuries', 'Total Serious Injuries', 'Total Fatal Injuries']

# change the datatype to int for the noted columns to be changed to integer
for c in int_cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')
    df[c] = df[c].astype(pd.Int64Dtype())

# get datatypes for all of the variables
df.dtypes
```

### Other Additional Data Manipulation Steps
There were a few other steps made here to further manipulate the data for usage. 
The data in the 'Make' column was changed to uppercase for consistency.
For the 'Injury Severity' column, the parentheses and text was removed after the fatal fields.
A 0 would replace NA in 'Total Fatal Injuries' column, if the injury severity in the same row was non-fatal.

```python
# make the data in the 'make' column all uppercase for consistancy
df['Make'] = df['Make'].str.upper()

# create a function that removes parentheses and text inside
def remove_parentheses_text(text):
    # find index of '('
    last_open_paren = text.rfind('(')
    # if '(' is found, remove the text
    if last_open_paren != -1:
        return text[:last_open_paren].strip()
    # if '(' is not found, return original text
    return text

# apply the remove parentheses function to the 'Injury Serverity' column
df['Injury Severity'] = df['Injury Severity'].apply(remove_parentheses_text)

# if injury severity is non-fatal fill a 0 for total fatal injuries
def replace_nonfatal_na(x):
    # check to see if injury severity is non-fatal
    if x['Injury Severity'] == 'Non-Fatal':
        # replace with 0 fatalities
        return 0
    else:
        # return the original <NA> value
        return x['Total Fatal Injuries']
    
# apply the function to the Total Fatal Injuries column
df['Total Fatal Injuries'] = df.apply(replace_nonfatal_na, axis=1)
```

## Question 2: 
(10 pts) Calculate the number of events that transpired each year and display the results using a line chart. Which year has the highest number of events?


```python
# find the number of events that happened each year
yearly_count = df.groupby(df["Event Date"].dt.year)["Event Date"].count()

# create a plot that visualizes the number of events that happpen each year
plt.plot(yearly_count.index, yearly_count, marker = 'o')
# give the plot a title
plt.title('Events per Year: Aviation Accidents and Incidents')
# label the x axis
plt.xlabel('Event Date (Year)')
# label the y axis
plt.ylabel('Event Count')

# show the line chart
plt.show()
```

Based on the line chart created above, it appears that the year with the highest number of events occured in the year 1982.

## Question 3
(10 pts) Create a suitable visualization that shows the proportion of Accidents and Incidents in the dataset. Hint: use the Investigation Type. Explain the results.

```python
# find the number of investigations classified as accidents
accident_count = (df["Investigation Type"] == "Accident").sum()
# find the number of investigations classified as incidents
incident_count = (df["Investigation Type"] == "Incident").sum()

# store the accident and incident counts, respectively
counts = [accident_count, incident_count]
# store the accident and incident category names, respectively
investigations = ['Acccidents', 'Incidents']

# create a bar graph containing the types of investigations and their count
plt.bar(investigations, counts)
# give the bar graph a title
plt.title('Investigation Types: Accidents and Incidents')
# label the x-axis to represent investigation type
plt.xlabel('Investigation Type')
# label the x-axis to represent the count of investigations
plt.ylabel('Count')

# show the plot
plt.show()
```

Based on the results it appears there is a much higher proportion of accidents compared to incidents. This could mean that a lot of aviation investigations are accidents rather than an intentional incident.

## Question 4
(10) Create a suitable visualization that calculates the number of events grouped by the aircraft make (e.g. Cessna, Boeing, Piper, etc).** 

Visualize which aircraft has the most events; only show the top ten results. Comment on the results.

```python
# calculate the number of events grouped by aircraft make
make_event_counts = df.groupby("Make")["Make"].count()

# get the top ten results
top_10_make_event_counts = make_event_counts.nlargest(10)

# create a bar graph showing aircraft makes and how many related events have occured
plt.bar(top_10_make_event_counts.index, top_10_make_event_counts)
# rotate the x-axis bar names to be readable
plt.xticks(rotation=45)
# give the bar graph a title
plt.title('Top 10 Event Counts Based on Aircraft Make')
# label the x-axis to be the make
plt.xlabel('Aircraft Make')
# label the y-axis to show the number of events
plt.ylabel('Number of Events')

# show the bar graph
plt.show()
```

Based on the results, Cessna has a much greater event count than any other make of aircraft. This is followed by Piper with a little over than half as many incidents. Third comes Beech with approximately 5000 events. After this, the rest of the top ten are fairly closely grouped together.

```python
# find the number of fatal and non-fatal events
# injury_count = df_boeing_737.groupby("Injury Severity")["Injury Severity"].count()

# find the number of events classified as fatal
fatal_count = (df_boeing_737["Injury Severity"] == "Fatal").sum()
# find the number of events classified as non-fatal
nonfatal_count = (df_boeing_737["Injury Severity"] == "Non-Fatal").sum()

# store the fatal and non-fatal counts, respectively
counts = [fatal_count, nonfatal_count]
# store the Fatal and Non-Fatal category names, respectively
injury_serverity = ['Fatal', 'Non-Fatal']

# create a bar chart for the top counts for event phases
plt.bar(injury_serverity, counts)
# give the plot a title
plt.title('Fatal and Non-Fatal Events for Boeing 737')
# label the x axis
plt.xlabel('Injury Serverity')
# label the y axis
plt.ylabel('Event Count')

# show the bar chart
plt.show()
```
